```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("data/delta_flyers_feed.xml")
root = tree.getroot()
```

```python
# Get the root and top-level tags
print("Root tag", root.tag)
for child in root:
    print(child.tag)
```

```python
# Inspect the channel to discover its structure
first_channel = root.find("channel")
```

```python
for elem in first_channel:
    print(elem.tag)
```

```python
for channel in root.findall('channel'):
    title = channel.find('title').text
    link = channel.find('link').text
    description = channel.find('description').text
    language = channel.find('language').text
    item = channel.find('item')
    print(f"Podcast: {title}")
    print(f"Link: {link}")
    print(f"Description: {description}")
    print(f"Language: {language}")
```

```python
# Inspect items
for item_elem in item:
    print(item_elem.tag)
```
